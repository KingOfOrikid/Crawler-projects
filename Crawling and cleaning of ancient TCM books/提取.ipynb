```python
path='普济方（儿科全部.txt'
f=open(path,'w',encoding='utf-8')
for key in dic.keys():
    f.write(key+'\n')
    f.write('\n')
    #卷
    for name in dic[key].keys():
        f.write(name+'\n')
        if '论' in dic[key][name].keys():
            item=dic[key][name]['论']
            if type(item)==list:
                for i in item:
                    f.write(i+'\n')
            else:
                f.write(item+'\n')
        f.write('\n')
        
        #卷里的药方
        if '药方' in dic[key][name].keys() and dic[key][name]['药方']!=[]:
            con=dic[key][name]['药方']
            for item in con:
                if '方名' in item:
                    f.write(item['方名']+'\n')
                if '主治' in item:
                    f.write(item['主治']+'\n')
                if '中药及其剂量' in item:
                    for zhong in item['中药及其剂量']:
                        f.write(zhong+' '+item['中药及其剂量'][zhong]+'；')
                    f.write('\n')
                if '用法' in item:
                    f.write(item['用法']+'\n')
                if '后药' in item:
                    for zhong in item['后药']:
                        f.write(zhong+' '+item['后药'][zhong]+'；')
                    f.write('\n')
                if '后药用法' in item:
                    f.write(item['后药用法']+'\n')
                if '又方' in item:
                    for you in item['又方']:
                        f.write('又方'+'\n')
                        if '方名' in you:
                            f.write(you['方名']+'\n')
                        if '主治' in you:
                            f.write(you['主治']+'\n')
                        if '中药及其剂量' in you:
                            for zhong in you['中药及其剂量']:
                                f.write(zhong+' '+you['中药及其剂量'][zhong]+'；')
                            f.write('\n')
                        if '用法' in you:
                            f.write(you['用法']+'\n')
                f.write('\n')
            f.write('\n')
            f.write('\n')
```
